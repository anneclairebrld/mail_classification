# Feature Engineering 
 ## Time stamp classification

```python
## **** Approach 1 - working day Monday to Friday, 8 AM to 6 PM

def flag_df(dframe):

    if (8 <= dframe['hour'] <= 18) and (dframe['weekday'] <= 5):
        return int(0)
    else:
        return int(1)

Dfnew['work_flag'] = Dfnew.apply(flag_df, axis = 1)
Dfnew2['work_flag'] = Dfnew2.apply(flag_df, axis = 1)


Dfnew['work_flag'].astype(str).astype(int)
Dfnew2['work_flag'].astype(str).astype(int)


## ********Create 3 buckets for time and 2 for day

def hour_flag(Dfnew):

    if (16 <= Dfnew['hour'] < 24):
        return int(0)
    elif(0 <= Dfnew['hour'] < 8):
        return int(1)
    elif(8 <= Dfnew['hour'] < 16):
        return int(2)

Dfnew['hour_flag'] = Dfnew.apply(hour_flag, axis = 1)
Dfnew2['hour_flag'] = Dfnew2.apply(hour_flag, axis = 1)

Dfnew['hour_flag'].astype(str).astype(int)
Dfnew2['hour_flag'].astype(str).astype(int)

#----------- DAY FLAG --------------
def day_flag(Dfnew):

    if (Dfnew['weekday'] <= 5):
        return int(0)
    else:
        return int(1)

Dfnew['day_flag'] = Dfnew.apply(day_flag, axis = 1)
Dfnew2['day_flag'] = Dfnew2.apply(day_flag, axis = 1)


Dfnew['day_flag'].astype(str).astype(int)
Dfnew2['day_flag'].astype(str).astype(int)
```

## Character Bucketing

```python
def char_subj_flag(Dfnew):

    if (Dfnew['chars_in_subject'] >= 150):
        return int(0)
    elif(70 <= Dfnew['chars_in_subject'] < 150 ):
        return int(1)
    elif(20 <= Dfnew['chars_in_subject'] < 70):
        return int(2)
    elif( Dfnew['chars_in_subject'] < 20):
        return int(3)

Dfnew['char_subj_flag'] = Dfnew.apply(char_subj_flag, axis = 1)
Dfnew2['char_subj_flag'] = Dfnew2.apply(char_subj_flag, axis = 1)

Dfnew['char_subj_flag'].astype(str).astype(int)
Dfnew2['char_subj_flag'].astype(str).astype(int)

# ------------CHAR IN BODY BUCKETING------------

def char_main_flag(Dfnew):

    if (Dfnew['chars_in_subject'] > 4000):
        return int(0)
    elif(4000 <= Dfnew['chars_in_subject'] < 3000 ):
        return int(1)
    elif(3000 <= Dfnew['chars_in_subject'] < 2000):
        return int(2)
    elif(2000 <= Dfnew['chars_in_subject'] < 1000):
        return int(3)
    elif( Dfnew['chars_in_subject'] < 1000):
        return int(4)

Dfnew['char_main_flag'] = Dfnew.apply(char_main_flag, axis = 1)
Dfnew2['char_main_flag'] = Dfnew2.apply(char_main_flag, axis = 1)

Dfnew['char_main_flag'].astype(str).astype(int)
Dfnew2['char_main_flag'].astype(str).astype(int)
```
